```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sylviarezk/Dataanalytics-2023/main/DOHMH_Dog_Bite_Data.csv')
```

```python
df.head(3)
```

```python
df['UniqueID'].is_unique
```

```python
df_breed_clean = df[df['Breed']!= 'UNKNOWN']
```

```python
df_breed_clean
```

```python
df_breed_clean.Species.unique()
```

```python
df_breed_grouped = df_breed_clean.groupby('Breed')['UniqueID'].count().sort_values(ascending=False)
```

```python
df_breed_20 = df_breed_grouped.head(20)
```

```python
df_breed_20.plot(kind='bar')
```

```python
df_pitbull = df_breed_clean[df_breed_clean['Breed']=='Pit Bull']
```

```python
df_pitbull.head(20)
```

```python
df_pitbull_borough_clean = df_pitbull[df_pitbull['Borough'] !='Other']
```

```python
df_pitbull_borough = df_pitbull_borough_clean.groupby('Borough')['UniqueID'].count().sort_values(ascending=False)
```

```python
df_pitbull_borough
```

```python
df_pitbull_borough.plot(kind='bar')
```

```python
df_pitbull_gender = df_pitbull_borough_clean.groupby('Gender')['UniqueID'].count().sort_values(ascending=False)
```

```python
df_pitbull_gender
```

```python
df_pitbull_gender.plot(kind='bar')
```
